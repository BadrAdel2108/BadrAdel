```python
def assign_holiday(date):
    if date in ['12-02-2010', '11-02-2011', '10-02-2012', '08-02-2013']:
        return 'Super Bowl'
    elif date in ['10-09-2010', '09-09-2011', '07-09-2012', '06-09-2013']:
        return 'Labour Day'
    elif date in ['26-11-2010', '25-11-2011', '23-11-2012', '29-11-2013']:
        return 'Thanksgiving'
    elif date in ['31-12-2010', '30-12-2011', '28-12-2012', '27-12-2013']:
        return 'Christmas'
    else:
        return 'Non-Holiday'
```

```python
holiday_analysis_df = data.drop(['Store', 'Year', 'Month'], axis='columns')
holiday_analysis_df['Holiday'] = holiday_analysis_df['Date'].apply(assign_holiday)
holiday_analysis_df.head()
```

```python
mean_df = holiday_analysis_df.groupby('Holiday').agg(Mean_Weekly_Sales=('Weekly_Sales','mean'))
mean_df
```

```python
mean_df.sort_values(by='Mean_Weekly_Sales')
```

```python
mean_df.plot(kind='bar', y='Mean_Weekly_Sales', figsize=(8,8))
```

```python
import calendar

fig = px.bar(data_frame = data.groupby('Month').sum().reset_index(),
             x = 'Month', y = 'Weekly_Sales',
             text = 'Weekly_Sales', height = 550)

fig.update_layout(title = ' Monthly Sales from 2011 to 2013',
                  yaxis_title = 'Total Sales',
                  )

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')

fig.update_yaxes(tickprefix="$")

fig.update_layout(
    xaxis = dict(
        ticktext = [calendar.month_name[n] for n in range(1,13)],
        tickvals = [n for n in range(1,13)]
    )
)

fig.show()
```

```python
fig = px.bar(data_frame = data.groupby(['Month','Year']).sum().reset_index(),
             x = 'Month', y = 'Weekly_Sales', color = 'Year',
             text = 'Weekly_Sales', height = 550)

fig.update_layout(title = 'Monthly Sales by Year',
                  yaxis_title = 'Total Sales',
                  )


fig.update(layout_coloraxis_showscale=False)

fig.update_yaxes(tickprefix="$")

fig.update_layout(
    xaxis = dict(
        ticktext = [calendar.month_name[n] for n in range(1,13)],
        tickvals = [n for n in range(1,13)]
    )
)

fig.show()
```
